```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/Excalibur/Documents/heatmap/year.csv")
```

```python
heatmap_data = pd.pivot_table(data, values='count', index=['month'], columns='year')
```

```python
heatmap_data.index = pd.CategoricalIndex(heatmap_data.index, categories=[
                                         "January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])
```

```python
heatmap_data.sortlevel(level=0, inplace=True)
fig, ax = plt.subplots()
fig.set_size_inches(14, 10)
sns.heatmap(heatmap_data, cmap="YlGnBu")
```

```python
acc_data = pd.read_csv("C:/Users/Excalibur/Documents/acc_analysis/acc_analysis.csv")
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)
sns.barplot(x="count", y="Weather_Conditions", hue="Accident_Severity", data=acc_data)
```

From the heatmap we can see that over the year most number of accidents happens during months July, October, and November. These months mostly fall on vacation periods.

```python
cas_data = pd.read_csv("C:/Users/Excalibur/Documents/cas_analysis/cas_analysis.csv")
fig, ax = plt.subplots()
fig.set_size_inches(10, 10)
sns.barplot(x="count", y="Casualty_Type", hue="Accident_Severity", data=cas_data)
```

we can see that most numbers of casualties include car occupants at first, pedestrians are second and Cyclist area at third. Car occupants take up almost 80% of the casualties with most injuries are just slight and not fatal.

```python
hour_data = pd.read_csv("C:/Users/Excalibur/Documents/hour_analysis/hour_analysis.csv")
fig, ax = plt.subplots()
fig.set_size_inches(15, 10)
sns.barplot(x="hour", y="count", data=hour_data)
```

we can see that most number of accidents occurs during 8 to 9 in the morning when people mostly commute to the office and also in the evening between 3 to 6 when mostly go back to home from work.
